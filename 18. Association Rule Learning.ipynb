```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.metrics import confusion_matrix
from sklearn.datasets import make_classification
from matplotlib.colors import ListedColormap
from apyori import apriori
import seaborn as sns
sns.set(style="white")

%matplotlib inline
```

```python
loc = "data/Part 5 - Association Rule Learning/Section 28 - Apriori/Python/"

dataset = pd.read_csv(loc + 'Market_Basket_Optimisation.csv', header=None)
dataset.head()
```

```python
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i, j]) for j in range (0, 20)])
```

```python
# Training Apriori on the dataset
# 3*7/7500
rules = apriori(transactions, min_support=0.003, min_confidence=.2, min_lift=3, min_length=2)
results = list(rules)
```

```python
for result in range(5):
    print(results[result])
    print()
```

```python
## Putting the results well organised into a Pandas DataFrame
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])

## Displaying the results non sorted
resultsinDataFrame

## Displaying the results sorted by descending lifts
resultsinDataFrame.nlargest(n = 10, columns = 'Lift')
```
